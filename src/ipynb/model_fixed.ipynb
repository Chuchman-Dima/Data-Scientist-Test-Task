# Feature Engineering

```python
# Переведемо часові фічі в потрібний нам формат
df_params['timestamp_reg'] = pd.to_datetime(
    df_params['timestamp_reg'],
    format='mixed'
)

df_actions['timestamp_interval_start'] = pd.to_datetime(
    df_actions['timestamp_interval_start'],
    format='mixed'
)

df_actions['timestamp_interval_end'] = pd.to_datetime(
    df_actions['timestamp_interval_end'],
    format='mixed'
)
```

```python
# Видалення дублікатів.
print("Duplicate: ", df_params.duplicated().sum())
print("Duplicate: ", df_actions.duplicated().sum())

df_actions = df_actions.drop_duplicates()
df_params = df_params.sort_values('timestamp_reg').drop_duplicates(subset='id_user', keep='first')

print("After delete duplicate: ", df_params.duplicated().sum())
print("After delete duplicate: ", df_actions.duplicated().sum())
print("Now lines in params:", df_params.shape[0])
```

```python
# Агрегуємо actions по юзеру
# Кожен юзер має 4 інтервали (6-годинні) => агрегуємо суми та розраховуємо поведінкові фічі

actions_agg = df_actions.groupby('id_user').agg(
    act_sum_payments=('sum_payments', 'sum'),
    act_max_payments=('sum_payments', 'max'),
    act_cnt_payments=('cnt_payments', 'sum'),
    act_sum_credits=('sum_credits_spend', 'sum'),
    act_cnt_returns=('cnt_returns', 'sum'),
    act_sum_visits=('cnt_visit_other_users', 'sum'),
    act_max_visits=('cnt_visit_other_users', 'max'),
    # Чи була активність взагалі (будь-яка дія) у кожному з інтервалів
    act_active_intervals=('cnt_visit_other_users', lambda x: (x > 0).sum()),
).reset_index()

print(actions_agg.shape)
actions_agg.head()
```

# Навчання моделей


```python
X = df.drop(columns=['id_user', 'revenue_7d', 'act_cnt_returns'])
y = df['revenue_7d']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
y_bin_train = y_train.gt(0).astype(int)
y_bin_test  = y_test.gt(0).astype(int)

print(f"Тренувальна вибірка: {X_train.shape[0]} рядків")
print(f"Тестова вибірка:     {X_test.shape[0]} рядків")
print(f"Платників у train:   {y_bin_train.sum()} ({y_bin_train.mean()*100:.1f}%)")
print(f"Платників у test:    {y_bin_test.sum()}  ({y_bin_test.mean()*100:.1f}%)")

# TimeSeriesSplit — CV без заглядання в майбутнє
tscv = TimeSeriesSplit(n_splits=3)
```

```python
clf_base = LGBMClassifier(
    objective='binary',
    random_state=42,
    n_jobs=-1
)

param_grid_clf = {
    'max_depth': [3, 5, 7],
    'num_leaves': [15, 31, 63],
    'min_child_samples': [50, 100],
    'scale_pos_weight': [10, 30, 48]
}

grid_clf = GridSearchCV(
    estimator=clf_base,
    param_grid=param_grid_clf,
    scoring='average_precision',
    cv=tscv,
    verbose=2,
    n_jobs=-1
)

grid_clf.fit(X_train, y_bin_train)

print(f"Найкращі параметри Класифікатора: {grid_clf.best_params_}")
p_payer_test = grid_clf.best_estimator_.predict_proba(X_test)[:, 1]
```

```python
payer_mask_train = y_train > 0
X_train_payers = X_train[payer_mask_train]
y_train_payers = y_train[payer_mask_train]

# Логарифмуємо таргет — це стабілізує дисперсію і допомагає моделі
y_train_payers_log = np.log1p(y_train_payers)

reg_base = LGBMRegressor(
    objective = 'regression',
    random_state = 42,
    n_jobs = -1
)

param_grid_reg = {
    'max_depth': [3, 4, 5],
    'num_leaves': [7, 15, 31],
    'min_child_samples': [50, 100, 200],
    'reg_lambda': [0.0, 1.0, 10.0]
}

# ВИПРАВЛЕНО: раніше тут стояв KFold(n_splits=3, shuffle=False). На відсортованих по часу
# даних це означало, що в частині фолдів модель тренувалась на ПІЗНІШИХ даних, а
# валідувалась на РАНІШИХ — пряме витікання майбутнього в оцінку якості під час GridSearchCV.
# Використовуємо TimeSeriesSplit, як і для класифікатора, щоб CV не заглядав у майбутнє.
kf_reg = TimeSeriesSplit(n_splits=3)

grid_reg = GridSearchCV(
    estimator = reg_base,
    param_grid = param_grid_reg,
    scoring = 'neg_mean_absolute_error',
    cv = kf_reg,
    verbose = 1,
    n_jobs = -1
)

# Навчаємо на ЛОГАРИФМОВАНИХ значеннях виручки
grid_reg.fit(X_train_payers, y_train_payers_log)
print(f"Найкращі параметри Регресора: {grid_reg.best_params_}")

# Прогнозуємо на тесті та повертаємо з log
# Регресор отримує весь X_test (включно з неплатниками) — це свідомо:
# при множенні final_pred = proba * pred_reg неплатники отримають малу proba,
# що занулить їхній внесок у фінальний прогноз.
predicted_amount_test_log = grid_reg.best_estimator_.predict(X_test)
predicted_amount_test = np.expm1(predicted_amount_test_log).clip(min=0)
```

```python
# Фінальний прогноз = P(заплатить) * E(скільки заплатить)
# Математичне очікування виручки для кожного юзера.
final_pred = p_payer_test * predicted_amount_test
```

# Оцінка моделей

**Класифікатор(чи заплатить користувач):**
Accuracy тут не підходить — при дисбалансі 98/2% модель, яка завжди передбачає "не платник", дасть 98% accuracy і буде марною. Тому:

* ROC-AUC — показує наскільки добре модель ранжує платників вище за неплатників, незалежно від порогу. Корисний для загальної оцінки якості, але при сильному дисбалансі може виглядати оптимістично навіть якщо precision по факту низький (через величезну кількість негативів у знаменнику FPR).
* PR-AUC (average_precision) — більш інформативний при сильному дисбалансі класів, бо фокусується саме на меншому класі (платниках): і precision, і recall рахуються відносно кількості реальних/передбачених позитивів, а не всього датасету.

**ВИПРАВЛЕНО:** `GridSearchCV` для класифікатора раніше оптимізував `scoring='roc_auc'`, хоча фінальна оцінка якості — про PR-AUC. Тепер `scoring='average_precision'`.

**Регресор (скільки заплатить):**

* MAE — інтерпретується в доларах, стійкий до викидів (великих платежів "китів"). Зручний для бізнесу: "в середньому помиляємось на $X".
* RMSE — чутливий до великих помилок, тому показує чи модель сильно промахується по "китах".
* R² тут не є основною метрикою, бо при великій кількості нулів у тестовій вибірці може давати викривлену картину.

```python
# Оцінка Класифікатора
roc_auc = roc_auc_score(y_bin_test, p_payer_test)
print(f"ROC-AUC: {roc_auc:.4f}")

pr_auc = average_precision_score(y_bin_test, p_payer_test)
print(f"PR-AUC:  {pr_auc:.4f}")

# Поріг 0.5 — довільний вибір за замовчуванням, а не підібраний під бізнес-метрику
y_pred_clf = (p_payer_test >= 0.5).astype(int)
print("\nClassification Report:")
print(classification_report(y_bin_test, y_pred_clf, target_names=["Не платник", "Платник"]))

fig, ax = plt.subplots(figsize=(7, 5))
PrecisionRecallDisplay.from_predictions(y_bin_test, p_payer_test, ax=ax)
ax.set_title(f"Precision-Recall Curve  (PR-AUC = {pr_auc:.4f})")
plt.tight_layout()
plt.show()
```

```python
# Оцінка Регресора (тільки на платниках тестової вибірки)

payer_mask_test = y_test > 0
y_true_payers = y_test[payer_mask_test]
y_pred_payers = predicted_amount_test[payer_mask_test]

mae = mean_absolute_error(y_true_payers, y_pred_payers)
rmse = np.sqrt(mean_squared_error(y_true_payers, y_pred_payers))
r2 = r2_score(y_true_payers, y_pred_payers)

print(f"MAE (середня абс. помилка в $): {mae:.2f}")
print(f"RMSE (чутливий до викидів): {rmse:.2f}")
print(f"R² (тільки на платниках): {r2:.4f}")
```

```python
# Оцінка фінального прогнозу (на всій тестовій вибірці)

mae_final = mean_absolute_error(y_test, final_pred)
rmse_final = np.sqrt(mean_squared_error(y_test, final_pred))

print(f"Фінальний MAE (весь тест): {mae_final:.2f}")
print(f"Фінальний RMSE (весь тест): {rmse_final:.2f}")

# Scatter: реальне vs прогнозоване (тільки на платниках)
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(y_true_payers, y_pred_payers, alpha=0.4, edgecolors="white", linewidths=0.3)
max_val = max(y_true_payers.max(), y_pred_payers.max())
ax.plot([0, max_val], [0, max_val], color="red", linestyle="--", label="Пряма")
ax.set_xlabel("Реальна виручка ($)")
ax.set_ylabel("Прогнозована виручка ($)")
ax.set_title("Реальне vs Прогнозоване (платники тесту)")
ax.legend()
plt.tight_layout()
plt.show()
```

# Feature Importance — LightGBM

LightGBM рахує важливість ознак за кількістю сплітів (splits) або сукупним приростом (gain). Використовуємо `gain` — він показує фактичний внесок у зменшення помилки.

```python
clf_best = grid_clf.best_estimator_
feature_names = X_train.columns.tolist()

# importance_type='gain' — фактичний внесок у зменшення помилки (а не кількість сплітів)
importance_clf = pd.DataFrame({
    "feature": feature_names,
    "importance": clf_best.booster_.feature_importance(importance_type='gain')
}).sort_values("importance", ascending=False)

plt.figure(figsize=(9, 6))
sns.barplot(data=importance_clf, x="importance", y="feature", hue="feature", palette="magma", legend=False)
plt.title("Feature Importance — Класифікатор (LightGBM)")
plt.xlabel("Важливість (gain)")
plt.ylabel("Ознака")
plt.tight_layout()
plt.show()
```

```python
reg_best = grid_reg.best_estimator_

# importance_type='gain' — фактичний внесок у зменшення помилки (а не кількість сплітів)
importance_reg = pd.DataFrame({
    "feature": feature_names,
    "importance": reg_best.booster_.feature_importance(importance_type='gain')
}).sort_values("importance", ascending=False)

plt.figure(figsize=(9, 6))
sns.barplot(data=importance_reg, x="importance", y="feature", hue="feature", palette="magma", legend=False)
plt.title("Feature Importance — Регресор (LightGBM)")
plt.xlabel("Важливість (gain)")
plt.ylabel("Ознака")
plt.tight_layout()
plt.show()
```

```python
# Класифікатор
explainer_clf = shap.TreeExplainer(grid_clf.best_estimator_)

X_test_sample = X_test.sample(n=100000, random_state=42)

import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    shap_values_clf = explainer_clf.shap_values(X_test_sample)

# Для бінарної класифікації shap іноді повертає список [neg_class, pos_class] —
# явно перевіряємо і беремо клас "платник" (1), щоб не наплутати, який клас показуємо.
if isinstance(shap_values_clf, list):
    shap_values_clf = shap_values_clf[1]

shap.summary_plot(shap_values_clf, X_test_sample)
```

```python
# Регресія
explainer_reg = shap.TreeExplainer(grid_reg.best_estimator_)

# Для регресора візьмемо лише платників
X_test_payers = X_test[payer_mask_test]
shap_values_reg = explainer_reg.shap_values(X_test_payers)

shap.summary_plot(shap_values_reg, X_test_payers)
```

# Обмеження рішення

- Поведінкові дані є лише за перший день (4 інтервали по 6 годин) — це дуже вузьке вікно для прогнозу виручки за 7 днів.
- Платників лише ~2% від усієї вибірки, тому регресор навчається на дуже малій кількості прикладів, що підвищує ризик перенавчання.
- Регресор навчається тільки на платниках, але при інференсі отримує всіх користувачів — розподіл фіч може відрізнятись.
- Поріг класифікатора (0.5) обраний за замовчуванням, а не підібраний під конкретну бізнес-метрику.

# Можливі покращення

- **Tweedie regression** як альтернатива двоетапному підходу — єдина модель, яка нативно моделює zero-inflated розподіл.
- Більше feature engineering: активність у конкретні інтервали (ранок/вечір), відношення `credits_spend / cnt_payments`.
- **Калібрування ймовірностей** класифікатора (`CalibratedClassifierCV`) для коректнішого математичного очікування при фінальному множенні `p * amount`.
- **Підбір оптимального порогу** класифікації під бізнес-метрику (наприклад, максимум F1 або очікуваного прибутку) замість фіксованих 0.5.
- Додати **валідаційну вибірку** окремо від тестової (train / val / test), щоб підбір гіперпараметрів не «бачив» тест жодного разу.

# Загальний висновок

**Що зроблено:** на основі даних реєстрації та поведінки користувача за перший день побудована модель, яка за кожним новим користувачем прогнозує очікувану виручку за наступні 7 днів.

### Як працює модель
З ~1.6 млн користувачів лише **2.1%** реально платять. Модель вирішує це у два кроки: спочатку визначає *чи заплатить* людина, потім — *скільки*. Фінальний прогноз — це добуток цих двох оцінок.

> Цифри нижче — з прогону після переходу `GridSearchCV` класифікатора на `scoring='average_precision'` (найкращі параметри: `max_depth=3, num_leaves=15, min_child_samples=100, scale_pos_weight=10`).

### Якість визначення платників
Модель правильно визначає **73%** реальних платників (recall), точність по платниках — 85% (precision) — це вище, ніж було на `roc_auc`-версії (73%), тобто оптимізація під PR-AUC дала менше хибних спрацювань ціною трохи нижчого recall. Загальна здатність розрізняти платників від неплатників: ROC-AUC = 0.956, PR-AUC = 0.801 — саме PR-AUC тепер оптимізується напряму, і 0.80 при базовій частці платників ~2% — це суттєво краще за випадкове вгадування (PR-AUC випадкової моделі дорівнював би ~0.02).

### Точність прогнозу суми
Для тих, хто реально платить, модель помиляється в оцінці суми в середньому на 32.98 (RMSE 136.88, R² = 0.47). При цьому медіанна виручка платника — 9.99 USD, а 75% платять до 26.49 USD — тобто помилка велика переважно через "китів" (є юзери з виручкою до 20 000 USD).

На **всій базі** (включно з неплатниками) середня помилка прогнозу — 1.15 (RMSE 19.31) — що логічно: більшість отримують прогноз ≈ 0 USD, і це правда.

### Що це дає маркетингу та продукту

**1. Ефективність сегментів (Висновки з EDA)**

* **По каналах залучення:**
  | Канал (traffic_type_id) | Середня виручка за 7д |
  |---|---|
  | 2 | 1.47 |
  | 3 | 1.05 |
  | 1 | 0.80 |
  Канал 2 приносить **вдвічі більше виручки** з одного користувача порівняно з каналом 1.

* **По пристроях:**
  | Платформа | Середня виручка |
  |---|---|
  | Desktop | 1.85 |
  | Tablet | 1.37 |
  | Mobile | 1.03 |
  Desktop-користувач коштує на **80% більше**, ніж мобільний.

* **По країнах** — топ за середньою виручкою: GP (10.1 USD), MC (9.4 USD), GU (5.4 USD), PF (4.3 USD), IS (3.4 USD). (NF з mean=64.98 не враховуємо — лише 1 юзер, статистично ненадійно).

**2. Ключові драйвери виручки (Інтерпретація моделі через графік SHAP)**

* **Прямі фінансові маркери першого дня (`act_cnt_payments`, `act_sum_payments`, `act_sum_credits_spend`):** орієнтовно найпотужніші предиктори — користувачі, які роблять кілька транзакцій або активно витрачають кредити в перший же день, генерують найбільшу виручку протягом наступного тижня.

* **Залученість (`act_active_intervals`):** активність користувача в різних 6-годинних інтервалах першого дня — сигнал для продуктової команди про важливість утримання уваги протягом усього першого дня.
